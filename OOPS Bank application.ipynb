```python
import re
class Bank:
  bankname='India Bank'
  branch='Thane,Maharashtra'

  #account creation
  def __init__(self,username,pancard,address,number):
    self.username=username
    self.pancard=pancard
    self.address=address
    self.balance=0.0
    self.number=number
    print(f'Hello {self.username} congrats your account has been created!!!!!!!')

  #deposit amount
  def cheque_deposit(self,amount,cheque_no):
    self.balance+=amount
    print(f'{amount} has been deposited with cheque number {cheque_no} successfully......')
  def cash_deposit(self,amount):
    self.balance+=amount
    print(f'{amount} has been deposited successfully......')
  def online_deposit(self,amount,upi):
    self.balance+=amount
    print(f'{amount} has been deposited with upi id {upi} successfully......')

  #withdraw amount
  def cash_withdraw(self,amount):
    if amount<self.balance:
      self.balance-=amount
      print(f'{amount} has been withdrawn successfully......')
    else:
      print('Sorry....Insufficient balance....')
  def cheque_withdraw(self,amount,cheque_no):
    if amount<self.balance:
      self.balance-=amount
      print(f'{amount} has been withdrawn with cheque number {cheque_no} successfully......')
    else:
      print('Sorry....Insufficient balance....')
  def online_withdraw(self,amount,upi):
    if amount<self.balance:
      self.balance-=amount
      print(f'{amount} has been withdrawn with upi id {upi} successfully......')
    else:
      print('Sorry....Insufficient balance....')

  #ministatement
  def ministatement(self):
    print(f'Your account balance is {self.balance}')


print(f'Welcome to {Bank.bankname},{Bank.branch} branch.........')
username=input("Enter your name:")
pancard=input("Enter pancard number:")
address=input("Enter your address:")
number=int(input('Enter your number:'))

#object created
b=Bank(username,pancard,address,number)

while True:
  print('Select any option:')
  print('1.Deposit\n2.Withdraw\n3.Ministatement\n4.Exit')
  option=int(input(' '))

  if option == 1:
    while True:
      print('Please select mode of deposit.......')
      print('1.Cheque\n2.Cash\n3.Online\n4.Main menu')
      mode=int(input(' '))
      if mode == 1:
        cheque_no=int(input('Enter cheque number:'))
        a=re.fullmatch("\d{6}",str(cheque_no))
        if a!=None:
          amount=float(input('Enter amount to be deposited:'))
          b.cheque_deposit(amount,cheque_no)
        else:
          print('Please enter valid cheque number....')
      if mode == 2:
        amount=float(input('Enter amount to be deposited:'))
        b.cash_deposit(amount)
      if mode == 3:
        upi=input("Enter upi Id(as number@upi):")
        c=re.fullmatch("[7-9]\d{9}@upi",str(upi))
        if c!=None:
          amount=float(input('Enter amount to be deposited:'))
          b.online_deposit(amount,upi)
        else:
          print('Please enter valid upi')
      if mode == 4:
        break
  
  if option == 2:
    while True:
      print('Please select mode of withdraw.......')
      print('1.Cheque\n2.Cash\n3.Online\n4.Main menu')
      mode=int(input(' '))
      if mode == 1:
        cheque_no=int(input('Enter cheque number:'))
        a=re.fullmatch("\d{6}",str(cheque_no))
        if a!=None:
          amount=float(input('Enter amount to be withdraw:'))
          b.cheque_withdraw(amount,cheque_no)
        else:
          print('Please enter valid cheque number....')
      if mode == 2:
        amount=float(input('Enter amount to be withdrawn:'))
        b.cash_withdraw(amount)
      if mode == 3:
        upi=input("Enter upi Id(as number@upi):")
        c=re.fullmatch("[7-9]\d{9}@upi",str(upi))
        if c!=None:
          amount=float(input('Enter amount to be withdrawn:'))
          b.online_withdraw(amount,upi)
        else:
          print('Please enter valid upi')
      if mode == 4:
        break
  
  if option == 3:
    b.ministatement()
  
  if option == 4:
    print(f'Thanks for using {Bank.bankname}.........\nPlease Visit again!!')
    break
```
